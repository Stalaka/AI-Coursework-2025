```python
#1. Write a program that asks the user for a number of days. The program then prints out the number of seconds in the amount of days given.
days = int(input("Enter number of days: "))
seconds = days * 24 * 60 * 60   # convert days to seconds
print(days, "days have", seconds, "seconds")
```

```python
#2. Write a program that asks a user to input the radius then the program calculates the volume of a sphere (the formula for the volume is (4/3)πr3).
import math   # import math library

r = float(input("Enter radius of the sphere: "))
volume = (4/3) * math.pi * (r ** 3)   # formula (4/3)*pi*r^3
print("The volume is:", volume)  # print result
```

```python
#6.Write a Python program that does the following:
#• Uses a loop for a user to continually input 5 values to populate an array.
#• Calculates and displays the average of the values input into the array.
nums = []   # create empty list

for i in range(5):   # loop 5 times
  n = float(input("Enter number " + str(i+1) + ": "))
  nums.append(n)

total = sum(nums)   # sum of list
avg = total / 5     # divide by 5
print("Numbers entered:", nums)
print("Average:", avg)
```
